```python
def loyalty_split(alldata):
    if alldata['Account Length']<=10:
        return 1
    elif (alldata['Account Length']>10) & (alldata['Account Length']<=20):
        return 2
    elif (alldata['Account Length']>20) & (alldata['Account Length']<=30):
        return 3
    elif (alldata['Account Length']>30) & (alldata['Account Length']<=40):
        return 4
    elif (alldata['Account Length']>40) & (alldata['Account Length']<=50):
        return 5
    elif (alldata['Account Length']>50) & (alldata['Account Length']<=100):
        return 6
    elif (alldata['Account Length']>100) & (alldata['Account Length']<=150):
        return 7
    elif (alldata['Account Length']>150) & (alldata['Account Length']<=200):
        return 8
    elif (alldata['Account Length']>200) & (alldata['Account Length']<=250):
        return 9
    elif alldata['Account Length']>250:
        return 10



alldata['Loyalty']=alldata.apply(lambda alldata:loyalty_split(alldata), axis=1)
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sb
import numpy as np
from scipy.stats import norm
from sklearn.preprocessing import StandardScaler
from scipy import stats
import warnings
warnings.filterwarnings('ignore')
%matplotlib inline

sb.distplot(alldata['CustServ Calls'], fit=norm);
fig = plt.figure()
res = stats.probplot(alldata['CustServ Calls'], plot=plt)
```

```python
X = alldata.loc[:,['International Plan','VMail Plan', 'Day Mins', 'International calls','CustServ Calls', 'Loyalty']]
y = alldata.Churn

from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X,y,
                                test_size=0.2,random_state=10)
```

```python
from sklearn.linear_model import LogisticRegression
model=LogisticRegression()
model.C=0.01
model.densify
model.fit(X_train,y_train)
y_predict=model.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score,confusion_matrix
print('Accuracy score:',accuracy_score(y_test,y_predict))
print('confuion matrix:\n',confusion_matrix(y_test,y_predict))
from sklearn.metrics import classification_report
print(classification_report(y_test,y_predict))
```

```python
from sklearn.neural_network import MLPClassifier
model=MLPClassifier(hidden_layer_sizes=(100,100),
                   learning_rate_init=0.009,
                   max_iter=1000,
                    random_state=4
                   )

model.fit(X_train,y_train)
y_predict = model.predict(X_test)        
y_train_predict = model.predict(X_train)
from sklearn.metrics import accuracy_score,confusion_matrix
print("Train accuracy :",accuracy_score(y_train,y_train_predict))
print("Test accuracy :",accuracy_score(y_test,y_predict))
```
